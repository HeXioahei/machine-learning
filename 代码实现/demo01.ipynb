```python
# 导入包
import numpy as np    # 取别名
import matplotlib.pyplot as plt   # 取别名

# 生成横轴数据
x_array = np.linspace(0, 2*np.pi, 100) 
# numpy.linspace(start, stop, num=50, endpoint=True)
# 生成在给定范围内的等差数列。
# start代表数值序列的起始值，stop代表结束值，num代表数值序列的数量。
# endpoint表示是否包含结束点。
# 即在[0, 2pi]内生成一个100个数值的等差数列。

# 正弦和余弦函数数据
sin_y = np.sin(x_array)
cos_y = np.cos(x_array)
# Numpy库中的数学函数，用于计算给定角度的正余弦值，
# 其参数可以是单个弧度值，也可以是一个数组

# 设置图片的大小
fig, ax = plt.subplots(figsize=(8, 6))
# 创建一个新的 Matplotlib 图形 fig 和 一个轴 ax 对象
# 并设置大小为（8， 6），单位为英寸

# 绘制正弦和余弦曲线
ax.plot(x_array, sin_y, label='sin', color='r', linewidth=2)
ax.plot(x_array, cos_y, label='cos', color='b', linewidth=2)
# 线宽的单位是点(point, pt)，通常用于测量线条、字体等绘图元素的大小。
# 默认情况下，一个点等于 1/72 inch。
# linewidth表示线宽，也可简作 lw。

# 设置标题
ax.set_title('Sine and cosine functions')
# 设置横轴和纵轴标签
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

# 添加图例，用于标识不同曲线或数据系列
ax.legend()

# 设置横轴和纵轴的范围
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)

# 定义横轴刻度的位置，从 0 到 2pi，间隔为 pi/2。
x_ticks = np.arange(0, 2*np.pi+np.pi/2, np.pi/2)
# 定义横轴刻度的标签
x_ticklabels = [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
# 设置刻度位置和标签
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels)

# 横纵轴采用相同的比例，保证图形在绘制时不会因为坐标轴的比例问题而产生形变。
ax.set_aspect('equal')
plt.grid()   # 启用网格线，默认参数为True.

# 将图片存成SVG格式
plt.savefig('正弦_余弦函数曲线.svg', format='svg')

# 显示图形
plt.show()
```

**创建图形、轴对象，添加子图，绘制曲线**

```python
import numpy as np    # 取别名
import matplotlib.pyplot as plt   # 取别名

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

fig = plt.figure()  # 用figure()生成一个Figure对象
ax = fig.add_subplot(2, 2, 1)  
ax2 = fig.add_subplot(2, 2, 2)  
ax3 = fig.add_subplot(223)  
ax4 = fig.add_subplot(2, 2, 4) 
# 用add_subplot()方法创建一个新的子图轴对象ax，并将其添加到Figure对象中
# 参数 2,2,1表示子图在 2 行 2 列的空间的第 1 个位置
# 注意，2,3,1也可以写成 221.

# 然后在子图中绘制一个新的正弦曲线
ax.plot(x, y)
ax2.plot(x, y)
ax3.plot(x, y)
ax4.plot(x, y)

# 设置标签以便区分
ax.set_title('sin1')
ax2.set_title('sin2')
ax3.set_title('sin3')
ax4.set_title('sin4')

y_sin = np.sin(x)
y_cos = np.cos(x)
x_ticks = np.arange(0, 2*np.pi+np.pi/2, np.pi)
x_ticklabels = [r'$0$', r'$\pi$', r'$2\pi$']

fig2, (axx1, axx2) = plt.subplots(1, 2, figsize=(10,4), sharey=True)
# 参数 1, 2 表示 1 行 2 列的子图布局。
# 参数 figsize=(10,4) 指定了图形的宽度和高度。
# 参数 sharey=True 表示两个子图共享如同的 y 轴，在垂直方向上由相同的刻度和范围
# 而 fig2, (axx1,axx2) 将 plt.subplots 的返回值解包，
# 其中 fig2 是整个图形对象，(axx1,axx2) 是一个包含两个子图对象的元组。

axx1.plot(x, y_sin, color='blue') # 绘制曲线，颜色为蓝色
axx1.set_title('Sine function')   # 设置标题
axx1.set_xlabel('x')              # 设置横轴标签
axx1.set_ylabel('f(x)', rotation='horizontal', ha='right')  # 设置纵轴标签
# rotation='horizontal'设置标签的旋转角度。'horizontal' 表示标签将水平放置。
# 也可以使用具体的角度值（如 0, 90, 180 等）来指定旋转角度。
# ha='right'设置标签的水平对齐方式。'right'表示标签将向右对齐。
# 其他可能的值包括 'center'（居中对齐）和 'left'（左对齐）。

axx1.set_xlim(0, 2*np.pi)  # 设置横轴刻度范围
axx1.set_ylim(-1.5, 1.5)   # 设置纵轴刻度范围
axx1.set_xticks(x_ticks)   # 设置横轴刻度位置
axx1.set_xticklabels(x_ticklabels)   # 设置横轴刻度标签
axx1.grid(True)            # 启用网格线
axx1.set_aspect('equal')

axx2.plot(x, y_cos, color='red')
axx2.set_title('Cosine function')
axx2.set_xlabel('x')
axx2.set_ylabel('f(x)', rotation='horizontal', ha='right')
axx2.set_xlim(0, 2*np.pi)
axx2.set_ylim(-1.5, 1.5)
axx2.set_xticks(x_ticks)
axx2.set_xticklabels(x_ticklabels)
axx2.grid(True)
axx2.set_aspect('equal')

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```

##### **使用 Plotly 绘制线图**
Plotly可以绘制具有交互属性的图形

```python
# 导入包
import plotly.express as px
import numpy as np
import pandas as pd

# 生成横轴数据
x = np.linspace(0,  2*np.pi, 100)

# 生成正弦和余弦曲线数据
y_sin = np.sin(x)
y_cos = np.cos(x)

# 生成 Pandas 数据帧
df = pd.DataFrame({'x':x, 'Sine':y_sin, 'Cosine':y_cos})
# 键是数据帧中的列名，而值是对应列的数据。

# 创建图表1
fig1 = px.line(x=x, y=[y_sin, y_cos], labels={'y':'f(x)', 'x':'x'})
# 用字典键值对来指定横纵轴的标签

# 创建图表2
fig2 = px.line(df)
# 直接以新创建的数据帧 df 为输入数据，直接以 数据帧列 为标签

# 修改图例，不然会为 'wide_variable_0'，'wide_variable_1'
fig1.data[0].name = 'Sine'
fig1.data[1].name = 'Cosine'
fig2.data[0].name = 'Sine'
fig2.data[1].name = 'Cosine'

# 显示图表
fig1.show()
fig2.show()
print(df)
```
