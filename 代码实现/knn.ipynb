寻找最近的K个数据，推测新数据的分类。

通用步骤：
计算距离-->升序排列-->取前K个-->加权平均

K的选取：不能太大，会导致分类模糊；不能太小，会受个例的影响，波动较大

如何选取K：根据经验或均方根误差


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split  

# 读取数据
iris = load_iris()
datas = iris.data

# print(iris)

# 分割训练集和测试集
X_train,X_test,y_train,y_test = train_test_split(iris['data'],iris['target'],random_state=0)
# print(X_train, len(X_train))
# print(X_test, len(X_test))
# print(y_train)
# print(y_test)

# KNN
# 距离
def distance(d1, d2):
    res = 0
    for i in range(0, 4):
        res += (float(d1[i])-float(d2[i]))**2
    res = res**0.5
    return res

# print(distance(X_train[0], X_train[1]))

K=7   # K不能太大也不能太小
def knn(data):
    #1。求距离
    res = [ {"result":y_train[i], "distance":distance(data, X_train[i])}
           for i in range(0,len(y_train)) ]
    # print(res)
    # print()

    #2。对距离进行排序
    res = sorted(res, key=lambda item:item['distance']) # 每个 item 就是res里的一个元素
    # print(res)
    # print()
    
    #3。取前 K 个进行加权平均
    res2 = res[0:K]   # 取前 K 个
    sum = 0   # 总距离
    weight = {'0':0, '1':0, '2':0}
    for r in res2:
        sum += r['distance']
    for r in res2:
        weight[str(r['result'])] += 1 - r['distance']/sum
    
    # print(weight)
    
    # 返回判断结果
    label='0'
    max = 0
    for i in weight.keys():
        if weight[i] > max:
            max = weight[i]
            label = i
    # print(label)
    return label

# knn(X_test[7])
# print(y_test[7])

# 测试该 knn 模型的正确率
def test(data_test, result_test):
    count = 0
    for data, result in zip(data_test, result_test):
        # print(knn(data), result)
        if knn(data) == str(result):
            count = count + 1

    # print()
    # print(count)
    return count/len(data_test)

# 进行测试
print(test(X_test, y_test))

# 进行预测
mydata = [9.4, 4.3, 7.3, 6.3]
print(knn(mydata))
```
