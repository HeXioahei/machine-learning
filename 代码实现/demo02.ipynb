#### 平面散点图
鸢尾花数据集是一个常用于机器学习和统计学习的示例数据集。
导入鸢尾花数据集的三个途径：
* sklearn.datasets.load_iris()，Numpy Array类型，特征数据和标签数据分别保存
* seaborn.load_dataset("iris")，Pandas DataFrame类型，数据帧的列表标签为'sepal_length','sepal_width','petal_length','petal_width','species'。
* plotly.express.data.iris()，Pandas DataFrame类型，'sepal_length','sepal_width','petal_length','petal_width','species','species_id'。'species_id'这一列的标签为整数0、1、2.

```python
# 导入包
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris  
import numpy as np

# 加载鸢尾花数据集
iris = load_iris()  
# 用 load_iris 函数加载经典的鸢尾花数据集，
# 它的返回类型是 scikit-learn 中一个简单的数据容器类

# 提取花萼长度和花萼宽度作为变量
sepal_length = iris.data[:,0] # 第1列，所有行。为花萼长度的样本数据。
sepal_width = iris.data[:,1] # 第2列，所有行。为花萼宽度样本数据。
target = iris.target # 提取鸢尾花数据集标签数组
#  iris.data 特征数据
# iris.target 标签数据

# 创建图形对象和轴对象
fig, ax = plt.subplots()

# 创建散点图 （横轴数据点，纵轴数据点，颜色由target数组决定，颜色映射）
plt.scatter(sepal_length, sepal_width, c=target, cmap='rainbow')
# target中的三个值（0，1，2）将通过'rainbow'映射到三个颜色，表达鸢尾花的三个类别。

# 添加标题和轴标签
ax.set_title('Iris sepal length vs width')
ax.set_xlabel('Sepal length (cm)')
ax.set_ylabel('Sepal width (cm)')

# 设置横纵轴刻度
ax.set_xticks(np.arange(4, 8+1, step=1))
ax.set_yticks(np.arange(1, 5+1, step=1))

# 设置横纵轴尺度1：1
ax.axis('scaled')

# 增加刻度网络，颜色为浅灰
# ax.grid(linestyle='--', linewidth=0.25, color=[0.7, 0.7, 0.7])
ax.grid(ls='--', lw=0.25, c=[0.7, 0.7, 0.7])

# 设置横纵轴范围
# ax.set_xbound(lower=4, upper=8)
# ax.set_ybound(lower=1, upper=5)
ax.set_xlim(4, 8)
ax.set_ylim(1, 5)

# 显示图形
plt.show()

```

##### Plotly

```python
# 导入包
import numpy as np
import plotly.express as px

# 从Plotly中导入鸢尾花样本数据，数据类型为 Pandas DataFrame
iris_df = px.data.iris()

# 绘制散点图，不渲染marker
fig = px.scatter(iris_df, x="sepal_length", y="sepal_width",
                 width=600, height=600, 
                 labels={"sepal_length":"Sepal length (cm)",
                         "sepal_width":"Sepal width (cm)"})
# 修饰图像
fig.update_layout(xaxis_range=[4,8], yaxis_range=[1,5])
xticks = np.arange(4,8+1)
yticks = np.arange(1,5+1)
fig.update_layout(xaxis=dict(tickmode='array',tickvals=xticks))
fig.update_layout(yaxis=dict(tickmode='array',tickvals=yticks))
fig.show()

# 绘制散点图，渲染 marker 展示鸢尾花分类
fig = px.scatter(iris_df, x="sepal_length", y="sepal_width",
                 color="species",
                 width=600, height=600, 
                 labels={"sepal_length":"Sepal length (cm)",
                         "sepal_width":"Sepal width (cm)"})
# 修饰图像
fig.update_layout(xaxis_range=[4,8], yaxis_range=[1,5])
fig.update_layout(xaxis=dict(tickmode='array',tickvals=xticks))
fig.update_layout(yaxis=dict(tickmode='array',tickvals=yticks))
fig.update_layout(legend=dict(yanchor='top',y=0.99,xanchor='left',x=0.01)) # 将图例位置调至左上角
fig.show()
```
